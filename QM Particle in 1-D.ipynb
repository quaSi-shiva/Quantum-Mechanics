```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.constants import h, m_e, e
import sympy as sp
import math as m
```

```python
x = np.linspace(0,2,1000)
def wavefunction(L,n,x):  
    F = ((2/L)**(1/2))*np.sin((n*np.pi*x)/L)
    return F
```

```python
fig,axs = plt.subplots(1,2,figsize=(10,3))

plt.subplots_adjust(hspace=0.4, wspace=0.2)

#wavefunction(L,n,x)

axs[0].plot(x,wavefunction(1,0,x), label = 'n = 0')
axs[0].plot(x,wavefunction(1,1,x), label = 'n = 1',lw = 0.8)
axs[0].plot(x,wavefunction(1,2,x), label = 'n = 2',lw=0.8)
#axs[0].plot(x,wavefunction(1,3,x), label = 'n = 3',lw=0.8)
axs[0].set_xlabel('L  ------>')
axs[0].set_ylabel('$\Psi~(x)$   ------>')
axs[0].set_title('$Particle~in~1-D~Box~of~length~L$')
axs[0].legend()
axs[0].grid()

axs[1].plot(x,(wavefunction(1,0,x)**2), label = 'n = 0',lw=0.8)
axs[1].plot(x,(wavefunction(1,1,x)**2), label = 'n = 1',lw=0.8)
axs[1].plot(x,(wavefunction(1,2,x)**2), label = 'n = 2',lw=0.8)
#axs[1].plot(x,(wavefunction(1,3,x)**2), label = 'n = 3',lw=0.8)
axs[1].set_xlabel('L  ------>')
axs[1].set_ylabel('$|\Psi|^2~(x)$   ------>')
axs[1].set_title('$Probability~Density~Curve$')
axs[1].legend(loc='upper right')
axs[1].grid()
```

```python
def Energy(n,L):
    Eev = ((h**2)/(8*m_e*L**2))/e
    return (n**2)*(Eev) #m=1
```

```python
y = list(np.linspace(0,2,100))

fig,axs = plt.subplots(2,2,figsize=(10,4))

plt.subplots_adjust(hspace=0.4, wspace=0.6)
plt.suptitle('Energy Level Diagrams for different width of boxes, L, L/2, 2L, L/4 respectively')

ev1,ev2,ev3,ev4 = [],[],[],[]

#L = 1

for i in range(len(y)):
    ev1.append(1*Energy(1,1))
    ev2.append(1*Energy(2,1))
    ev3.append(1*Energy(3,1))
    ev4.append(1*Energy(4,1))

axs[0,0].plot(y,ev1)
axs[0,0].plot(y,ev2)
axs[0,0].plot(y,ev3)
axs[0,0].plot(y,ev4)
axs[0,0].set_ylabel('Energy ----->')


#L = 1/2

L2ev1,L2ev2,L2ev3,L2ev4 = [],[],[],[]
for i in range(len(y)):
    L2ev1.append(1*Energy(1,2))
    L2ev2.append(1*Energy(2,2))
    L2ev3.append(1*Energy(3,2))
    L2ev4.append(1*Energy(4,2))

axs[0,1].plot(y,L2ev1)
axs[0,1].plot(y,L2ev2)
axs[0,1].plot(y,L2ev3)
axs[0,1].plot(y,L2ev4)
axs[0,1].set_ylabel('Energy ----->')


#L=2L

L3ev1,L3ev2,L3ev3,L3ev4 = [],[],[],[]
for i in range(len(y)):
    L3ev1.append(1*Energy(1,3))
    L3ev2.append(1*Energy(2,3))
    L3ev3.append(1*Energy(3,3))
    L3ev4.append(1*Energy(4,3))

axs[1,0].plot(y,L3ev1)
axs[1,0].plot(y,L3ev2)
axs[1,0].plot(y,L3ev3)
axs[1,0].plot(y,L3ev4)
axs[1,0].set_ylabel('Energy ----->')


#L = L/4

L4ev1,L4ev2,L4ev3,L4ev4 = [],[],[],[]
for i in range(len(y)):
    L4ev1.append(1*Energy(1,4))
    L4ev2.append(1*Energy(2,4))
    L4ev3.append(1*Energy(3,4))
    L4ev4.append(1*Energy(4,4))

axs[1,1].plot(y,L4ev1)
axs[1,1].plot(y,L4ev2)
axs[1,1].plot(y,L4ev3)
axs[1,1].plot(y,L4ev4)
axs[1,1].set_ylabel('Energy ----->')

None
```
